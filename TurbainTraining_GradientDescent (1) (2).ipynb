###  Batch Gradient Descent Without Feature Scaling......... with a learning rate=0.0000001 and 1500 iterations to converge the function for optimal coefficents....

```python
def Gradient_Descent(data):
    m=[0,0,0,0]
    c=0
    learning_rate=0.0001
    no_of_iterations=3000
    for i in range(no_of_iterations):
        m,c=Step_Gradient(data,m,c,learning_rate)
        print("The  m is {} and c is {} and cost {}".format(m,c,cost(data,m,c)))
    return m,c
```

```python
def cost(data,m,c):
    M=len(data)
    cost=0
    for i in range(M):
        
        x=np.array(data.iloc[i][['Temperature','Ambient_pressure','Relative_humidity','Exhaust_vaccume']])
        y=np.array(data.iloc[i][['Electrical_output']])
        cost=cost+(1/M)*(y-np.sum(m*x)-c)**2
    return cost
    
```

```python
def Step_Gradient(data,m,c,learning_rate):
    m_slope=[0,0,0,0]
    c_slope=0
    M=len(data)
    for i in range(len(data)):
        x=np.array(data.iloc[i][['Temperature','Ambient_pressure','Relative_humidity','Exhaust_vaccume']])
        y=np.array(data.iloc[i][['Electrical_output']])
        m_slope=m_slope+(-2/M)*(y-np.sum(m*x)-c)*x
        c_slope=c_slope+(-2/M)*(y-np.sum(m*x)-c)
    m=m-learning_rate*m_slope
    c=c-learning_rate*c_slope
    return m,c       
```

## Clearly from the results we can observe the cost function is decreasing which means i escaped from the problem of Overshooting so my parameters of learning rate and no of iterations seems to be appropriate here.But converging to optima may take more number of iterations.Learning reat choosen is 0.0000001 and no of iterations=3000.

```python
m,c=Gradient_Descent(df)
print("Final m {} and c {} are".format(m,c))
```
